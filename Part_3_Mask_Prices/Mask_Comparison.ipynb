```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2 Weddingstar
Weddingstar = "Resources/2-Weddingstar.csv"
Weddingstar_df = pd.read_csv(Weddingstar, encoding = 'utf-8')
#sort by date
Sort2=Weddingstar_df.sort_values(by='Date', ascending=True)
#convert to datetime
Sort2['Date'] = pd.to_datetime(Sort2['Date'])
#sort again
weddingstar_df = Sort2.sort_values(by='Date', ascending=True)
#plot
x_axis = weddingstar_df['Date']
y_axis = weddingstar_df['Price']
plt.plot(x_axis, y_axis)

#change
max = weddingstar_df['Price'].max()
min = weddingstar_df['Price'].min()
max_var = round(max-min, 0)
now = weddingstar_df['Price'][0]
length = len(weddingstar_df)
last = weddingstar_df['Price'][length-1]
per_change = round((100*((now-last)/last)), 0)
if per_change>0:
    color='g'
else:
    color='r'
plt.annotate(per_change, color=color, xy=(weddingstar_df['Date'][45], max-0.5))
plt.annotate('Percent Change = %', color='b', xy=(weddingstar_df['Date'][275], max-0.5))
plt.annotate('Max Price Var = $', color='b', xy=(weddingstar_df['Date'][275], max-1))
plt.annotate(max_var, color=color, xy=(weddingstar_df['Date'][50], max-1))

plt.xlabel("Time")
plt.ylabel("Price $")
plt.title("Weddingstar")
plt.xticks(rotation = 45)
plt.savefig("figures/weddingstar.png")
plt.show()
```
